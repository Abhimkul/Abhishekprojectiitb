```python
import numpy as np
import scipy
import scipy.interpolate
import matplotlib.pyplot as plot
```

```python
#Importing the Data: R12 O4
data  = np.loadtxt(r"C:\Users\Abhimkul\Downloads\sfhx_Final_Result (1)\sfhx_Final_Result\12\O4\kde_12_O4.txt") 
Rs = np.linspace(10.68344335 , 13.07864533,199)
kde = data[:,1]
```
