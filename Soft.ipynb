```python
import numpy as np
import scipy.stats 
from scipy.stats import gaussian_kde as kde
from scipy.stats import norm
import matplotlib.pyplot as plot
import random
import mplcyberpunk 

#Soft
np.random.seed(11)

A = norm(loc = 10, scale = 1)
B = A.rvs(size = 1000)
samples1 = np.sort(B)
#samples

x1 =  np.linspace(6,15,1000)
kernel1 = kde(samples1)
kde1 = kernel1.pdf(x1)

plot.plot(x1,kde1 , label = "soft" , color = "red" )
plot.xlabel("R")
plot.ylabel("p(R)")
plot.legend()
mplcyberpunk.add_glow_effects()
```
